```python
def evaluate(vti, individual):
    
    # split model into parts
    model = OrderedDict()
    nparams = vti.jac.nparams
    i0 = 0
    for part in nparams:
        i1 = i0 + nparams[part]
        
        model[part] = individual[i0:i1]
        
        i0 = i1
        
    # swap in trial model and calculate RMS
    model0 = copy.deepcopy(vti.fwd.model)
    vti.fwd.model = model
    rms = vti.fwd.rms
    vti.fwd.model = model0
    
    return 1./rms
```

```python
def generate_model(nparam, amin, amax):
    
    return amin + (amax - amin) * np.random.rand(nparam)

def rand(amin, amax):
    
    return amin + (amax - amin) * np.random.rand(1)
    
    
```

```python
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", np.ndarray, fitness=creator.FitnessMax)

toolbox = base.Toolbox()


_nparams = vti.jac.nparams
nparam = sum([_nparams[k] for k in _nparams])

toolbox.register("rand", rand, nparams, -5, 5)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.rand, nparam)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

toolbox.register("evaluate", evaluate)
toolbox.register("mate", tools.cxTwoPoints)
toolbox.register("mutate", tools.mutFlipBit, indpb=0.05)
toolbox.register("select", tools.selTournament, tournsize=3)

population = toolbox.population(n=10)
```
