```python
import numpy as np
import matplotlib.pyplot as plt
from rockfish.genetic.genetic import Population, Evolver
import bitstring
import copy
```

```python
def evaluate(ind):
    
    e = np.zeros(len(ind))
    for i in range(len(ind)):
        e[i] = np.sqrt((ind[i][0] + 1.)**2 + (ind[i][1] - 1.)**2)
        
    return 1. / e
```

```python
f0 = 5 * 2 * (np.random.rand(10, 2) - 0.5)

ga = Evolver(f0, evaluate, pm=0.8, px=0.8)

fig = plt.figure()
ax = fig.add_subplot(111)
#plt.xlim(, 20)
#plt.ylim(-20, 20)

ax.plot(ga.individuals[:, 0], ga.individuals[:, 1], '.k')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
#plt.xlim(, 20)
#plt.ylim(-20, 20)

ax.plot(ga.individuals[:, 0], ga.individuals[:, 1], '.k')

for i in range(100):
    
    ga.evolve(ngen=1, verbose=False)
    
    ax.plot(ga.individuals[:, 0], ga.individuals[:, 1], '.')
    
    if ga.best_fit >= 10000:
        break

ax.plot(ga.best_individual[0], ga.best_individual[1], '*r', markersize=20)

print i, ga.best_fit, ga.best_individual
```
